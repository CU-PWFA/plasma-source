```python
import sys
import os
sys.path.insert(0, "../../python")

import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as const
eps0 = const.epsilon_0
```

```python
# To find the fields from a Gaussian beam in the rest frame we use a multipole expansion
Q = -1e-9 #Coulombs
sigmar = 5.
sigmaz = 8.

x = np.linspace(-1e2, 1e2, 1000)
z = np.linspace(-1e2, 1e2, 1000)
x = x[:, None]
z = z[None, :]
r = np.sqrt(x**2 + z**2)
r[r < 10] = 10

dsigma2 = sigmaz**2 - sigmar**2
phi = Q*(1 + dsigma2*(2*z**2-x**2)/(2*r**4)) / (4*np.pi*eps0*r)

plt.imshow(np.flipud(np.transpose(phi)), aspect='auto', extent=[-1e2, 1e2, -1e2, 1e2])
cb = plt.colorbar()
cb.set_label('Potential (MV)')
plt.title('Potential from a Gaussian beam')
plt.xlabel(r'x ($\mu m$)')
plt.ylabel(r'z ($\mu m$)')
plt.show()
```
