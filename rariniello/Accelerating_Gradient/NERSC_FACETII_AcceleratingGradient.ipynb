```python
import sys
sys.path.insert(0, "../../python")

import os
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
plt.style.use('notes')
from scipy.optimize import curve_fit
import scipy.constants as const
me = const.physical_constants['electron mass energy equivalent in MeV'][0]
c = const.physical_constants['speed of light in vacuum'][0]

pathPre = '/home/robert/CU-PWFA/emittance_preservation/simulations/WitnessDelay/'
```

```python
beam_sigmaz = np.array([5.2, 7.8])
witness_delay = np.array([125., 135.,145., 155.])
gradient = np.zeros((len(beam_sigmaz), len(witness_delay)))
```

```python
data = np.load(pathPre + '125umX5.2um/AcceleratingGradient.npy').item()
gradient[0, 0] = np.average(data['gradient'][5:]/1e3)

plt.figure(figsize=(6,4), dpi=150)
gs = gridspec.GridSpec(2, 2)
plt.subplot(gs[0, 0])
plt.plot(data['x']*1e6, data['energy']/1e3)
plt.ylabel('Energy (GeV)')
plt.xlabel(r'z ($\mu m$)')
plt.twinx()
plt.plot(data['x'][1:]*1e6, data['gradient']/1e3, 'm')
plt.ylabel('Gradient (GeV/m)')

plt.subplot(gs[0, 1])
for i in range(len(data['x'])):
    plt.plot(data['xfield']*1e6, data['field'][i, :]/1e9, label=('%0.2f' % (data['x'][i]*1e6)))
#plt.legend()
plt.ylabel('Field (GeV/m)')
plt.xlabel(r'z ($\mu m$)')

plt.subplot(gs[1, 0])
plt.plot(data['x']*1e6, data['gamma_rms'])
plt.ylabel('Energy spread (%)')
plt.tight_layout()
plt.show()
```
