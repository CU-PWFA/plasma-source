```python
# Possible incoming beams for an adiabatic focuser
N = 3500
beta0 = 1
alpha0 = 0
gb0 = 20000
ne0 = 2*gb0/5.95074e4**2
dgdz0 = 0.0
a = 0.25
z = np.linspace(0, 100, N)
frac = 1/(1+a*z)**2
ne = frac*ne0
beta, alpha, gamma, gb = ecalc.cs_propagation(z, ne, beta0, alpha0, gb0, dgdz0, ne0)

Delta = a/(a-2)

plt.figure(figsize=(5, 3), dpi=150)
plt.plot(z, beta)
plt.plot(z, (1+a*z)*(1+1/(4/a**2-1)-1/(4/a**2-1)*np.cos(np.log(1+a*z)*np.sqrt(4/a**2-1))
                     -1/np.sqrt(4/a**2-1)*np.sin(np.log(1+a*z)*np.sqrt(4/a**2-1))), 'b--')
plt.plot(z, 1/np.sqrt(frac), 'm')
plt.fill_between(z, (1-Delta)/np.sqrt(frac), (1+Delta)/np.sqrt(frac), facecolor='m', alpha=0.25)
plt.title(r'Unmatched beam')
plt.xlabel(r'$z k_\beta$')
plt.ylabel(r'$\beta k_\beta$')
plt.legend([r'$\beta_{sim}$', r'$\beta_a$', r'$\beta_M$'], loc=9)
plt.twinx()
plt.plot(z, frac, 'g')
plt.text(75, 0.1, r'$\frac{n_e}{n_{e0}}=\frac{1}{(1+az)^2}$', color='green', fontsize=10)
plt.ylim(-0.05, 1.05)
plt.ylabel(r'$n_{e}/n_{e0}$')
plt.show()
```

```python
N = 3500
beta0 = 1.1
alpha0 = -.5
gb0 = 20000
ne0 = 2*gb0/5.95074e4**2
dgdz0 = 0.0
a = 0.2
z = np.linspace(0, 100, N)
frac = 1/(1+(a*z)**2)
ne = frac*ne0
beta, alpha, gamma, gb = ecalc.cs_propagation(z, ne, beta0, alpha0, gb0, dgdz0, ne0)

dz = z[1]-z[0]
bm = 1/np.sqrt(frac)
am=0
g0 = (1 + alpha0**2)/beta0
theta = np.cumsum(1/bm)*dz
analytic = bm/2*((1-g0)*np.cos(2*theta)+1+g0)-bm*alpha0*np.sin(2*theta)
analytic = bm/2*((beta0-g0+2*alpha0*am)*np.cos(2*theta)+beta0+g0-2*alpha0*am)-bm*(alpha0-beta0*am)*np.sin(2*theta)


plt.figure(figsize=(5, 3), dpi=150)
plt.plot(z, beta)
plt.plot(z, analytic, 'r--')
plt.plot(z, 1/np.sqrt(frac), 'm')
plt.title(r'Adiabaticity parameter')
plt.xlabel(r'$z k_\beta$')
plt.ylabel(r'$\beta k_\beta$')
plt.legend([r'$\beta_{sim}$', r'$\beta_{adi}$', r'$\beta_M$'], loc=9)
plt.twinx()
plt.plot(z, frac, 'g')
plt.text(75, 0.1, r'$\frac{n_e}{n_{e0}}=\frac{1}{1+(az)^2}$', color='green', fontsize=10)
plt.ylim(-0.05, 1.05)
plt.legend([r'$n_{e}/n_{e0}$', 'A'], loc=(0.1, .825))
plt.show()
```

```python
# Compare the adiabatic solution for the non physical linear ramp
N = 3500
beta0 = 1.1
alpha0 = 0.5
gb0 = 20000
ne0 = 2*gb0/5.95074e4**2
dgdz0 = 0.0
a = 0.2
z = np.linspace(0, 100, N)
frac = 1/(1+a*z)**2
ne = frac*ne0
beta, alpha, gamma, gb = ecalc.cs_propagation(z, ne, beta0, alpha0, gb0, dgdz0, ne0)

dz = z[1]-z[0]
#alpha0 = alpha0+beta0*a/2
am = -a/2
g0 = (1 + alpha0**2)/beta0
bm = 1/np.sqrt(frac)
theta = np.cumsum(1/bm)*dz
analytic = bm/2*((beta0-g0)*np.cos(2*theta)+beta0+g0)-bm*alpha0*np.sin(2*theta)

analytic = bm/2*((beta0-g0+2*alpha0*am)*np.cos(2*theta)+beta0+g0-2*alpha0*am)-bm*(alpha0-beta0*am)*np.sin(2*theta)

plt.figure(figsize=(5, 3), dpi=150)
plt.plot(z, beta)
plt.plot(z, analytic, 'r--', zorder=5)
plt.plot(z, 1/np.sqrt(frac), 'm')
plt.title(r'Adiabaticity parameter')
plt.xlabel(r'$z k_\beta$')
plt.ylabel(r'$\beta k_\beta$')
plt.legend([r'$\beta_{sim}$', r'$\beta_{adi}$', r'$\beta_M$'], loc=9)
plt.twinx()
plt.plot(z, frac, 'g')
plt.text(75, 0.1, r'$\frac{n_e}{n_{e0}}=\frac{1}{(1+az)^2}$', color='green', fontsize=10)
plt.ylim(-0.05, 1.05)
plt.legend([r'$n_{e}/n_{e0}$', 'A'], loc=(0.1, .825))
plt.show()
```
