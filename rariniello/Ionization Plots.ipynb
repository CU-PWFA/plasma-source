```python
import sys
sys.path.insert(0, "../python")

from ionization import ionization
from ionization import adk
import numpy as np
import matplotlib.pyplot as plt

# Test plotting
I = np.arange(1, 1500) * 0.01
H = adk.ionization_frac(13.5984, I, 100, 1, 0, 0)*100
He = adk.ionization_frac(24.5874, I, 100, 1, 0, 0)*100
Ar = adk.ionization_frac(15.7596, I, 100, 1, 0, 0)*100
Ar2 = adk.ionization_frac(27.62965, I, 100, 1, 0, 0)*100

plt.plot(I, H, 'b-', I, Ar, 'm-', I, He, 'g-', I, Ar2, 'r-')
plt.ylabel('Ionization %')
plt.xlabel('Intensity ($10^{14}\,\mathrm{W/cm}^2$)')
plt.title('Ionization percentage vs. laser intensity')
plt.legend(['H','Ar','He','Ar$^+$'])
plt.show()
```

```python
IUpper = I * 1.01
ILower = I * 0.99
# Hydrogen
HUpper = (adk.ionization_frac(13.5984, IUpper, 100, 1, 0, 0)*100 / H - 1)*100
HLower = (adk.ionization_frac(13.5984, ILower, 100, 1, 0, 0)*100 / H - 1)*100
# Helium
HeUpper = (adk.ionization_frac(24.5874, IUpper, 100, 1, 0, 0)*100 / He - 1)*100
HeLower = (adk.ionization_frac(24.5874, ILower, 100, 1, 0, 0)*100 / He - 1)*100
# Argon
ArUpper = (adk.ionization_frac(15.7596, IUpper, 100, 1, 0, 0)*100 / Ar - 1)*100
ArLower = (adk.ionization_frac(15.7596, ILower, 100, 1, 0, 0)*100 / Ar - 1)*100
# Argon +1
Ar2Upper = (adk.ionization_frac(27.62965, IUpper, 100, 1, 0, 0)*100 / Ar2 - 1)*100
Ar2Lower = (adk.ionization_frac(27.62965, ILower, 100, 1, 0, 0)*100 / Ar2 - 1)*100

plt.figure(1, figsize=(10,6))
plt.suptitle('% change in ionization fraction for 1% intensity variation')
plt.subplots_adjust(hspace=.25, wspace=0.5)

plt.subplot(221)
plt.plot(I, HUpper, 'b-', I, HLower, 'r-')
plt.ylabel('Ionization %$\Delta$')
#plt.xlabel('Intensity ($10^{14}\,\mathrm{W/cm}^2$)')
plt.title('Hydrogen')
plt.xlim([0.5,1.5])
plt.ylim([-10,10])
plt.legend(['% Increase','% Decrease'])
plt.twinx()
plt.plot(I, H, 'k-')
plt.xlim([0.5,1.5])
plt.ylabel('Ionization %')

plt.subplot(222)
plt.plot(I, HeUpper, 'b-', I, HeLower, 'r-')
plt.ylabel('Ionization %$\Delta$')
#plt.xlabel('Intensity ($10^{14}\,\mathrm{W/cm}^2$)')
plt.title('Helium')
plt.xlim([3,10])
plt.ylim([-10,10])
plt.legend(['% Increase','% Decrease'])
plt.twinx()
plt.plot(I, He, 'k-')
plt.xlim([3,10])
plt.ylabel('Ionization %')

plt.subplot(223)
plt.plot(I, ArUpper, 'b-', I, ArLower, 'r-')
plt.ylabel('Ionization %$\Delta$')
plt.xlabel('Intensity ($10^{14}\,\mathrm{W/cm}^2$)')
plt.title('Argon')
plt.xlim([0.8,2.2])
plt.ylim([-10,10])
plt.legend(['% Increase','% Decrease'])
plt.twinx()
plt.plot(I, Ar, 'k-')
plt.xlim([0.8,2.2])
plt.ylabel('Ionization %')

plt.subplot(224)
plt.plot(I, Ar2Upper, 'b-', I, Ar2Lower, 'r-')
plt.ylabel('Ionization %$\Delta$')
plt.xlabel('Intensity ($10^{14}\,\mathrm{W/cm}^2$)')
plt.title('Argon$^+$')
plt.xlim([5,15])
plt.ylim([-10,10])
plt.legend(['% Increase','% Decrease'])
plt.twinx()
plt.plot(I, Ar2, 'k-')
plt.xlim([5,15])
plt.ylabel('Ionization %')

plt.show()
```
