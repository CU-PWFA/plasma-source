```python
import sys
sys.path.insert(0, "../../python")

import numpy as np
import matplotlib.pyplot as plt
from ht import fht
```

```python
# Test of the FHT on some known functions
N = 1000
r = np.linspace(0, 100, N)
f = np.exp(-0.5*r**2)

k, h = fht.fht(r, f)
hexact = np.exp(-0.5*k**2)

plt.plot(k, np.real(h), 'b-', k, hexact, 'm-')
plt.title(r'FHT of a Gaussian')
plt.xlabel('k')
plt.ylabel('h(k)')
plt.legend(['FHT', 'Exact'])
plt.xlim([0, 10])
plt.ylim([-.1, 1.25])
plt.show()
```

```python
# IHT test of the same thing
r2, f2 = fht.ifht(k, h)
plt.plot(r2, np.real(f2), 'b-', r, f, 'm-')
plt.title(r'IFHT(FHT) of a Gaussian')
plt.xlabel('r')
plt.ylabel('f(r)')
plt.legend(['IFHT(FHT)', 'Original'])
plt.xlim([0, 10])
plt.ylim([-.1, 1.25])
plt.show()
```
